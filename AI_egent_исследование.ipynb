# Попытка в мультиагентную систему

```python
class Agent:
    """
    Базовый класс агента, реализующий метод решения задачи с использованием набора инструментов.
    """
    def __init__(self, name):
        self.name = name

    def generate_response(self, task):
        """
        Симуляция работы инструмента агента.
        Здесь можно интегрировать вызов нейросетевой модели или другой библиотеки.
        """
        # Простой пример: возвращаем текст с отметкой, что задача решена
        return f"{self.name} решил задачу: {task}"

    def solve_task(self, task):
        """
        Метод для решения задачи.
        В базовой реализации агент просто решает задачу самостоятельно.
        """
        result = self.generate_response(task)
        return result


class SubAgent(Agent):
    """
    Агент, который может иметь подчинённых для делегирования задач.
    Если подчинённые отсутствуют, решает задачу самостоятельно.
    """
    def __init__(self, name, sub_agents=None):
        super().__init__(name)
        self.sub_agents = sub_agents if sub_agents is not None else []

    def solve_task(self, task):
        if self.sub_agents:
            print(f"{self.name}: Имеет подчинённых, делегирую задачу: '{task}'")
            # Пример разбиения задачи на подзадачи: разбиваем по запятой
            subtasks = [s.strip() for s in task.split(',') if s.strip()]
            results = []
            # Делегируем каждую подзадачу по очереди подчинённым агентам
            for i, subtask in enumerate(subtasks):
                agent = self.sub_agents[i % len(self.sub_agents)]
                print(f"{self.name}: Передаю подзадачу '{subtask}' агенту {agent.name}")
                res = agent.solve_task(subtask)
                results.append(res)
            # Объединяем результаты подзадач
            combined_result = " ".join(results)
            return combined_result
        else:
            # Если подчинённых нет, решаем самостоятельно
            return super().solve_task(task)


class BossAgent(Agent):
    """
    Главный агент (boss), который получает исходную задачу, делит её на подзадачи
    и делегирует их своим подчинённым.
    """
    def __init__(self, name, sub_agents, verifier):
        super().__init__(name)
        self.sub_agents = sub_agents  # Список агентов для делегирования
        self.verifier = verifier      # Агент, проверяющий правильность решения

    def solve_task(self, task):
        print(f"{self.name}: Получена исходная задача: '{task}'")
        # Пример разбиения задачи: разбиваем по точкам
        subtasks = [s.strip() for s in task.split('.') if s.strip()]
        results = []
        # Делегируем каждую подзадачу по очереди подчинённым агентам
        for i, subtask in enumerate(subtasks):
            agent = self.sub_agents[i % len(self.sub_agents)]
            print(f"{self.name}: Делегирую подзадачу '{subtask}' агенту {agent.name}")
            res = agent.solve_task(subtask)
            results.append(res)
        # Объединяем результаты подзадач
        combined_result = " ".join(results)
        # Передаём объединённый результат на проверку
        if self.verifier.verify(combined_result):
            print(f"{self.name}: Результат проверен: {combined_result}")
            return combined_result
        else:
            print(f"{self.name}: Проверка результата не пройдена!")
            return "Ошибка в решении задачи"


class VerificationAgent:
    """
    Агент для проверки правильности решения.
    В данном примере проверка выполняется простым поиском ключевого слова.
    """
    def __init__(self, name):
        self.name = name

    def verify(self, result):
        print(f"{self.name}: Производится проверка результата...")
        # Пример проверки: если в результате присутствует строка 'решил задачу', считаем, что всё нормально
        if "решил задачу" in result:
            print(f"{self.name}: Проверка пройдена")
            return True
        else:
            print(f"{self.name}: Проверка не пройдена")
            return False


# Пример использования системы
if __name__ == "__main__":
    # Создаем агента проверки
    verifier = VerificationAgent("Verifier")

    # Создаем нижестоящих агентов (без подчинённых)
    worker1 = SubAgent("Worker1")
    worker2 = SubAgent("Worker2")

    # Создаем агента среднего уровня, который может делегировать задачу
    sub_leader = SubAgent("SubLeader", sub_agents=[worker1, worker2])

    # Создаем Boss агента, у которого в качестве подчинённых выступают sub_leader и другие рабочие агенты
    boss = BossAgent("Boss", sub_agents=[sub_leader, worker1, worker2], verifier=verifier)

    # Исходная задача для системы
    task = ("Проведи анализ трендов в нейросетях. "
            "Определи основные направления развития. "
            "Сделай прогноз на будущее.")

    final_result = boss.solve_task(task)
    print("\nИтоговый результат системы:")
    print(final_result)
```

```python
import together

# Класс LLM агента
class DeepSeekAgent:
    def __init__(self, model_name="deepseek-ai/DeepSeek-R1-Distill-Llama-70B-free"):
        self.client = together.Together()
        self.model = model_name

    def generate_response(self, user_input, max_tokens=512, temperature=0.7):
        try:
            response = self.client.chat.completions.create(
                model=self.model,
                messages=[{"role": "user", "content": user_input}],
                max_tokens=max_tokens,
                temperature=temperature
            )
            return response.choices[0].message.content
        except Exception as e:
            return f"Ошибка: {str(e)}"

# Базовый агент с локальной памятью
class Agent:
    def __init__(self, name, llm=None):
        self.name = name
        self.llm = llm if llm else DeepSeekAgent()
        self.local_memory = []

    def remember(self, info):
        self.local_memory.append(info)

    def solve_task(self, task):
        self.remember(f"Задача: {task}")
        prompt = f"{self.name} решает: {task}"
        result = self.llm.generate_response(prompt)
        self.remember(f"Результат: {result}")
        return result

# Подчинённый агент с возможностью делегирования
class SubAgent(Agent):
    def __init__(self, name, llm=None, sub_agents=None):
        super().__init__(name, llm=llm)
        self.sub_agents = sub_agents or []

    def solve_task(self, task):
        self.remember(f"Получена задача: {task}")
        if self.sub_agents:
            subtasks = [s.strip() for s in task.split(',') if s.strip()]
            results = []
            for i, subtask in enumerate(subtasks):
                agent = self.sub_agents[i % len(self.sub_agents)]
                result = agent.solve_task(subtask)
                results.append(result)
            combined = ' '.join(results)
            self.remember(f"Комбинированный результат: {combined}")
            return combined
        return super().solve_task(task)

# Агент верификации
class VerificationAgent:
    def __init__(self, name, llm=None):
        self.name = name
        self.llm = llm if llm else DeepSeekAgent()

    def verify(self, result):
        prompt = f"Проверь следующий результат на корректность: {result}. Ответь 'да' или 'нет'."
        answer = self.llm.generate_response(prompt).lower()
        return 'да' in answer, answer

# Главный агент с глобальной памятью и отчётом
class BossAgent(Agent):
    def __init__(self, name, llm=None, sub_agents=None, verifier=None):
        super().__init__(name, llm=llm)
        self.sub_agents = sub_agents or []
        self.verifier = verifier
        self.global_memory = {}

    def solve_task(self, task):
        print(f"{self.name}: Получена задача: '{task}'")
        subtasks = [s.strip() for s in task.split('.') if s.strip()]
        results = []
        for i, subtask in enumerate(subtasks):
            agent = self.sub_agents[i % len(self.sub_agents)]
            self.global_memory[subtask] = {'assigned_to': agent.name}
            result = agent.solve_task(subtask)
            self.global_memory[subtask]['result'] = result
            results.append(result)

        combined_result = ' '.join(results)
        print(f"{self.name}: Объединённый результат: {combined_result}\n")

        if self.verifier:
            passed, feedback = self.verifier.verify(combined_result)
            self.global_memory['verification'] = {'passed': passed, 'feedback': feedback}
            if passed:
                print(f"{self.name}: Проверка пройдена. Ответ пользователю:\n{combined_result}")
                return combined_result
            else:
                print(f"{self.name}: Проверка не пройдена. Комментарий: {feedback}")
                return f"Решение требует доработки. Комментарий проверяющего: {feedback}"
        return combined_result

# Пример использования
if __name__ == "__main__":
    shared_llm = DeepSeekAgent()
    verifier = VerificationAgent("Verifier", llm=shared_llm)

    worker1 = SubAgent("Worker1", llm=shared_llm)
    worker2 = SubAgent("Worker2", llm=shared_llm)
    sub_leader = SubAgent("SubLeader", llm=shared_llm, sub_agents=[worker1, worker2])

    boss = BossAgent("Boss", llm=shared_llm, sub_agents=[sub_leader, worker1, worker2], verifier=verifier)

    task = ("Проведи анализ трендов в нейросетях. "
            "Определи основные направления развития. "
            "Сделай прогноз на будущее.")

    final_result = boss.solve_task(task)
    print("\nГлобальная память Boss:")
    for k, v in boss.global_memory.items():
        print(f"{k}: {v}")
```

```python
import together
import datetime

# Класс LLM агента
class DeepSeekAgent:
    def __init__(self, model_name="deepseek-ai/DeepSeek-R1-Distill-Llama-70B-free"):
        self.client = together.Together()
        self.model = model_name

    def generate_response(self, user_input, max_tokens=512, temperature=0.7):
        try:
            response = self.client.chat.completions.create(
                model=self.model,
                messages=[{"role": "user", "content": user_input}],
                max_tokens=max_tokens,
                temperature=temperature
            )
            return response.choices[0].message.content
        except Exception as e:
            return f"Ошибка: {str(e)}"

# Утилита логирования
class Logger:
    def __init__(self, filename="agent_logs.txt"):
        self.filename = filename

    def log(self, message):
        timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        with open(self.filename, "a", encoding="utf-8") as f:
            f.write(f"[{timestamp}] {message}\n")
        print(message)

logger = Logger()

# Базовый агент с локальной памятью
class Agent:
    def __init__(self, name, llm=None):
        self.name = name
        self.llm = llm if llm else DeepSeekAgent()
        self.local_memory = []

    def remember(self, info):
        self.local_memory.append(info)

    def solve_task(self, task):
        self.remember(f"Задача: {task}")
        prompt = f"{self.name} решает: {task}"
        result = self.llm.generate_response(prompt)
        self.remember(f"Результат: {result}")
        logger.log(f"{self.name} решил: '{task}' -> {result}")
        return result

# Подчинённый агент с возможностью делегирования
class SubAgent(Agent):
    def __init__(self, name, llm=None, sub_agents=None):
        super().__init__(name, llm=llm)
        self.sub_agents = sub_agents or []

    def solve_task(self, task):
        self.remember(f"Получена задача: {task}")
        if self.sub_agents:
            subtasks = [s.strip() for s in task.split(',') if s.strip()]
            results = []
            for i, subtask in enumerate(subtasks):
                agent = self.sub_agents[i % len(self.sub_agents)]
                logger.log(f"{self.name}: делегирует '{subtask}' агенту {agent.name}")
                result = agent.solve_task(subtask)
                results.append(result)
            combined = ' '.join(results)
            self.remember(f"Комбинированный результат: {combined}")
            return combined
        return super().solve_task(task)

# Агент верификации
class VerificationAgent:
    def __init__(self, name, llm=None):
        self.name = name
        self.llm = llm if llm else DeepSeekAgent()

    def verify(self, result):
        prompt = f"Проверь следующий результат на корректность: {result}. Ответь 'да' или 'нет'."
        answer = self.llm.generate_response(prompt).lower()
        return 'да' in answer, answer

# Главный агент с глобальной памятью и отчётом
class BossAgent(Agent):
    def __init__(self, name, llm=None, sub_agents=None, verifier=None):
        super().__init__(name, llm=llm)
        self.sub_agents = sub_agents or []
        self.verifier = verifier
        self.global_memory = {}

    def solve_task(self, task, retry_on_fail=True):
        logger.log(f"{self.name}: Получена задача: '{task}'")
        subtasks = [s.strip() for s in task.split('.') if s.strip()]
        results = []

        for i, subtask in enumerate(subtasks):
            agent = self.sub_agents[i % len(self.sub_agents)]
            logger.log(f"{self.name}: Делегирую подзадачу '{subtask}' агенту {agent.name}")
            self.global_memory[subtask] = {'assigned_to': agent.name}
            result = agent.solve_task(subtask)
            self.global_memory[subtask]['result'] = result
            results.append(result)

        combined_result = ' '.join(results)
        logger.log(f"{self.name}: Объединённый результат: {combined_result}\n")

        if self.verifier:
            passed, feedback = self.verifier.verify(combined_result)
            self.global_memory['verification'] = {'passed': passed, 'feedback': feedback}

            if passed:
                logger.log(f"{self.name}: Проверка пройдена. Ответ пользователю:\n{combined_result}")
                return combined_result
            else:
                logger.log(f"{self.name}: Проверка не пройдена. Комментарий: {feedback}")
                if retry_on_fail:
                    logger.log(f"{self.name}: Перепоручаю задачу другим агентам")
                    rotated_agents = self.sub_agents[1:] + self.sub_agents[:1]
                    new_boss = BossAgent(self.name, self.llm, rotated_agents, self.verifier)
                    return new_boss.solve_task(task, retry_on_fail=False)
                return f"Решение требует доработки. Комментарий проверяющего: {feedback}"

        return combined_result

# Пример использования
if __name__ == "__main__":
    shared_llm = DeepSeekAgent()
    verifier = VerificationAgent("Verifier", llm=shared_llm)

    worker1 = SubAgent("Worker1", llm=shared_llm)
    worker2 = SubAgent("Worker2", llm=shared_llm)
    sub_leader = SubAgent("SubLeader", llm=shared_llm, sub_agents=[worker1, worker2])

    boss = BossAgent("Boss", llm=shared_llm, sub_agents=[sub_leader, worker1, worker2], verifier=verifier)

    task = ("Сочени песню про древних русов и ящеров")

    final_result = boss.solve_task(task)
    print("\nГлобальная память Boss:")
    for k, v in boss.global_memory.items():
        print(f"{k}: {v}")
```

```python
import together
import datetime
import numpy as np
from functools import lru_cache
from sentence_transformers import SentenceTransformer
from typing import List, Dict, Tuple

# Инициализация базовых компонентов
together.api_key = DEEPSEEK_API_KEY
encoder_model = SentenceTransformer('paraphrase-multilingual-mpnet-base-v2')

# --- Базовые классы ---
class DeepSeekAgent:
    """Модернизированный LLM-агент с кэшированием"""
    def __init__(self, model_name="deepseek-ai/DeepSeek-R1-Distill-Llama-70B-free"):
        self.client = together.Together()
        self.model = model_name

    @lru_cache(maxsize=512)
    def generate_response(self, user_input: str, max_tokens: int = 512, temperature: float = 0.7) -> str:
        try:
            response = self.client.chat.completions.create(
                model=self.model,
                messages=[{"role": "user", "content": user_input}],
                max_tokens=max_tokens,
                temperature=temperature
            )
            return response.choices[0].message.content
        except Exception as e:
            return f"Ошибка: {str(e)}"

class Tool:
    """Класс для работы с инструментами"""
    def __init__(self, name: str, func: callable, description: str):
        self.name = name
        self.func = func
        self.description = description

class Memory:
    """Векторная память с семантическим поиском"""
    def __init__(self):
        self.memories = []
        self.encoder = encoder_model

    def add(self, text: str) -> None:
        embedding = self.encoder.encode(text)
        self.memories.append({'text': text, 'embedding': embedding})

    def search(self, query: str, top_k: int = 3) -> List[Dict]:
        query_emb = self.encoder.encode(query)
        similarities = []
        for mem in self.memories:
            sim = np.dot(query_emb, mem['embedding'])
            similarities.append((mem['text'], sim))
        return sorted(similarities, key=lambda x: x[1], reverse=True)[:top_k]

class SecurityLayer:
    """Система безопасности и контроля"""
    def __init__(self):
        self.policies = [self._content_policy]

    def _content_policy(self, text: str) -> bool:
        forbidden = ["конфиденциально", "секретно", "паспортные данные"]
        return not any(fword in text.lower() for fword in forbidden)

    def validate(self, text: str) -> Tuple[bool, str]:
        for policy in self.policies:
            if not policy(text):
                return False, "Нарушение политики безопасности"
        return True, "OK"

# --- Усовершенствованные агенты ---
class Agent:
    """Базовый агент с расширенными возможностями"""
    def __init__(self, name: str, llm: DeepSeekAgent = None, tools: List[Tool] = None):
        self.name = name
        self.llm = llm or DeepSeekAgent()
        self.memory = Memory()
        self.tools = {t.name: t for t in (tools or [])}
        self.security = SecurityLayer()
        self.skills = {}

    def add_skill(self, skill: str, confidence: float) -> None:
        self.skills[skill] = confidence

    def evaluate_competence(self, task: str) -> float:
        prompt = f"""Оцени свою способность решить задачу от 0 до 1. Ответь только числом.
        Задача: {task}
        Навыки: {', '.join(self.skills.keys())}"""
        try:
            return float(self.llm.generate_response(prompt))
        except:
            return 0.0

    def use_tool(self, tool_name: str, input_data: str) -> str:
        if tool_name in self.tools:
            return self.tools[tool_name].func(input_data)
        return "Инструмент не найден"

    def solve_task(self, task: str) -> str:
        # Проверка безопасности
        is_valid, msg = self.security.validate(task)
        if not is_valid:
            return msg

        # Поиск в памяти
        similar = self.memory.search(task)
        if similar and similar[0][1] > 0.85:
            return f"Использую сохранённое решение: {similar[0][0]}"

        # Выбор инструмента
        if self.tools:
            tool_prompt = f"""Выбери инструмент для задачи. Доступно: {[t.name for t in self.tools.values()]}
            Задача: {task} Ответ в формате: ИНСТРУМЕНТ|ВВОД"""
            tool_response = self.llm.generate_response(tool_prompt)
            if '|' in tool_response:
                tool, input_data = tool_response.split('|', 1)
                return self.use_tool(tool.strip(), input_data.strip())

        # LLM-решение
        prompt = f"{self.name} решает задачу (отвечай на русском): {task}"
        result = self.llm.generate_response(prompt)
        self.memory.add(f"Задача: {task}\nРешение: {result}")
        return result

    def health_check(self) -> Dict:
        return {
            "llm": self._test_llm(),
            "tools": {name: "OK" for name in self.tools},
            "memory": f"{len(self.memory.memories)} записей"
        }

    def _test_llm(self) -> str:
        try:
            test = self.llm.generate_response("Тест связи", max_tokens=1)
            return "OK" if test else "ERROR"
        except:
            return "ERROR"

class TaskDecomposer:
    """Мультистратегическое разбиение задач"""
    STRATEGIES = {
        'sequential': "Разбей на последовательные шаги:",
        'parallel': "Разбей на независимые подзадачи:",
        'expert': "Определи необходимые области экспертизы:"
    }

    def __init__(self, llm: DeepSeekAgent):
        self.llm = llm

    def decompose(self, task: str, strategy: str = 'auto') -> Dict[str, str]:
        if strategy == 'auto':
            prompt = f"Выбери лучший метод разбиения (sequential/parallel/expert): {task}"
            strategy = self.llm.generate_response(prompt).lower()[:3]

        prompt = f"{self.STRATEGIES.get(strategy, 'Разбей на подзадачи:')}\n{task}"
        response = self.llm.generate_response(prompt)
        return self._parse_subtasks(response)

    def _parse_subtasks(self, text: str) -> Dict[str, str]:
        tasks = {}
        for line in text.split('\n'):
            if '.' in line:
                num, desc = line.split('.', 1)
                tasks[num.strip()] = desc.strip()
        return tasks

class BossAgent(Agent):
    """Усовершенствованный координатор"""
    def __init__(self, name: str, team: List[Agent], **kwargs):
        super().__init__(name, **kwargs)
        self.team = team
        self.decomposer = TaskDecomposer(self.llm)
        self.verifier = Agent("Verifier", llm=self.llm)

    def solve_task(self, task: str, retries: int = 2) -> str:
        # Разбиение задачи
        subtasks = self.decomposer.decompose(task)

        # Распределение задач
        results = {}
        for key, subtask in subtasks.items():
            agent = self.select_agent(subtask)
            results[key] = agent.solve_task(subtask)
            self.memory.add(f"Подзадача {key}: {results[key]}")

        # Компоновка результатов
        combined = "\n".join([f"{k}: {v}" for k, v in results.items()])

        # Верификация
        verification = self.verifier.solve_task(
            f"Проверь корректность:\n{combined}\nЗадача: {task}"
        )

        if "ошибка" in verification.lower() and retries > 0:
            return self._retry_mechanism(task, retries)

        return f"Итоговый отчет:\n{combined}\nПроверка: {verification}"

    def select_agent(self, task: str) -> Agent:
        scores = {agent.name: agent.evaluate_competence(task) for agent in self.team}
        return max(self.team, key=lambda a: scores[a.name])

    def _retry_mechanism(self, task: str, retries: int) -> str:
        self.team = self.team[1:] + self.team[:1]  # Ротация агентов
        return self.solve_task(task, retries-1)

# --- Пример использования ---
if __name__ == "__main__":
    # Инициализация инструментов
    def calculate(expr: str) -> str:
        try: return str(eval(expr))
        except: return "Ошибка вычисления"

    def web_search(query: str) -> str:
        # Фиктивная реализация для примера
        return f"Результаты по '{query}': 1. ... 2. ..."

    tools = [
        Tool("math", calculate, "Математические вычисления"),
        Tool("search", web_search, "Поиск информации")
    ]

    # Создание команды
    llm = DeepSeekAgent()
    analyst = Agent("Аналитик", llm=llm, tools=tools)
    engineer = Agent("Инженер", llm=llm, tools=[tools[0]])
    researcher = Agent("Исследователь", llm=llm, tools=[tools[1]])

    # Настройка навыков
    analyst.add_skill("анализ данных", 0.9)
    engineer.add_skill("оптимизация", 0.95)
    researcher.add_skill("исследования", 0.85)

    # Создание системы
    boss = BossAgent("Главный", team=[analyst, engineer, researcher], llm=llm)

    # Комплексная задача
    task = """
    Проанализировать рынок AI-чипов:
    1. Сравнить производительность NVIDIA H100 и AMD MI300X
    2. Рассчитать стоимость вычислений для 1ExaFLOP
    3. Выявить ключевые тренды на 2024-2025 гг.
    """

    result = boss.solve_task(task)
    print("Результат работы системы:")
    print(result)
    print("\nДиагностика:", boss.health_check())
```

```python
# @title Текст заголовка по умолчанию

import os
import io
import time
import json
import together
import datetime
import numpy as np
import pandas as pd
from google.oauth2.credentials import Credentials
from googleapiclient.discovery import build
from googleapiclient.http import MediaIoBaseUpload
from functools import lru_cache
from sentence_transformers import SentenceTransformer
from typing import List, Dict, Tuple, Optional

# Конфигурация
GOOGLE_CREDS_PATH = "path/to/your/credentials.json"
COLAB_NOTEBOOK_ID = "your-colab-notebook-id"
ENCODER_MODEL = SentenceTransformer('paraphrase-multilingual-mpnet-base-v2')

# Инициализация Together
together.api_key = "your-together-api-key"

class ColabIntegration:

    def __init__(self):
        self.creds = self._authenticate()
        self.service = build('colab', 'v1', credentials=self.creds)

    def _authenticate(self):
        scopes = ['https://www.googleapis.com/auth/cloud-platform',
                 'https://www.googleapis.com/auth/drive',
                 'https://www.googleapis.com/auth/colab']
        return Credentials.from_authorized_user_file(GOOGLE_CREDS_PATH, scopes)

    def execute_code(self, code: str, timeout: int = 300) -> dict:

        request = {
            'code': code,
            'timeout': f"{timeout}s"
        }

        try:
            execution = self.service.notebooks().executions().create(
                parent=f"notebooks/{COLAB_NOTEBOOK_ID}",
                body=request
            ).execute()

            execution_id = execution['name'].split('/')[-1]
            return self._poll_execution(execution_id, timeout)

        except Exception as e:
            return {'status': 'error', 'message': str(e)}

    def _poll_execution(self, execution_id: str, timeout: int) -> dict:

        start_time = time.time()
        while time.time() - start_time < timeout:
            result = self.service.notebooks().executions().get(
                name=f"notebooks/{COLAB_NOTEBOOK_ID}/executions/{execution_id}"
            ).execute()

            if result['state'] == 'SUCCEEDED':
                return {
                    'status': 'success',
                    'output': self._get_output(result['outputUri'])
                }
            elif result['state'] in ['FAILED', 'CANCELLED']:
                return {'status': 'error', 'message': result['errorMessage']}

            time.sleep(5)
        return {'status': 'timeout'}

    def _get_output(self, output_uri: str) -> dict:

        # Реализация загрузки результатов из Google Storage
        # (требуется дополнительная настройка)
        return {'raw_output': output_uri}

class CodeAnalysisTool:

    @staticmethod
    def analyze_output(output: dict) -> dict:
        analysis = {}

        if 'data' in output:
            try:
                df = pd.read_csv(io.StringIO(output['data']))
                analysis['stats'] = {
                    'rows': len(df),
                    'columns': list(df.columns),
                    'summary': df.describe().to_dict()
                }
            except Exception as e:
                analysis['error'] = f"Data analysis failed: {str(e)}"

        if 'plots' in output:
            analysis['plots_analysis'] = self._analyze_images(output['plots'])

        return analysis

    def _analyze_images(self, images: list) -> dict:
        # Реализация анализа изображений с помощью CV
        return {"images_count": len(images)}

class EnhancedLogger:

    def __init__(self, session_id: str = None):
        self.session_id = session_id or datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
        self.log_file = f"session_{self.session_id}.log"
        self.colab = ColabIntegration()

    def log(self, agent: str, action: str, data: dict):
        entry = {
            'timestamp': datetime.datetime.now().isoformat(),
            'agent': agent,
            'action': action,
            'data': data
        }

        self._write_to_file(entry)
        self._sync_to_colab()

    def _write_to_file(self, entry: dict):
        with open(self.log_file, 'a', encoding='utf-8') as f:
            f.write(json.dumps(entry, ensure_ascii=False) + '\n')

    def _sync_to_colab(self):
        try:
            with open(self.log_file, 'r') as f:
                content = f.read()

            media = MediaIoBaseUpload(io.BytesIO(content.encode()),
                                      mimetype='text/plain')
            self.colab.service.files().update(
                fileId=COLAB_NOTEBOOK_ID,
                media_body=media
            ).execute()
        except Exception as e:
            print(f"Sync error: {str(e)}")

class CodeGenerationAgent:
    """Агент для генерации и выполнения кода"""
    def __init__(self, llm, logger: EnhancedLogger):
        self.llm = llm
        self.logger = logger
        self.colab = ColabIntegration()
        self.analyzer = CodeAnalysisTool()

    def generate_and_execute(self, task: str) -> dict:
        self.logger.log(self.__class__.__name__, 'code_generation_start', {'task': task})

        # Генерация кода
        code = self._generate_code(task)
        self.logger.log(self.__class__.__name__, 'code_generated', {'code': code})

        # Выполнение кода
        execution_result = self.colab.execute_code(code)
        self.logger.log(self.__class__.__name__, 'code_executed', execution_result)

        # Анализ результатов
        if execution_result['status'] == 'success':
            analysis = self.analyzer.analyze_output(execution_result['output'])
            self.logger.log(self.__class__.__name__, 'analysis_complete', analysis)
            return {**execution_result, **analysis}

        return execution_result

    def _generate_code(self, task: str) -> str:
        prompt = f"""Сгенерируй Python код для решения задачи. Требования:
        - Используй только стандартные библиотеки и pandas/matplotlib
        - Добавь комментарии на русском языке
        - Сохрани данные в переменную df
        - Создай графики с помощью matplotlib
        - Верни результат в формате JSON

        Задача: {task}"""

        return self.llm.generate_response(prompt)

class ResearchAgent:
    """Агент для комплексных исследований"""
    def __init__(self, llm, logger: EnhancedLogger):
        self.llm = llm
        self.logger = logger
        self.code_agent = CodeGenerationAgent(llm, logger)
        self.memory = []

    def conduct_research(self, topic: str) -> dict:
        self.logger.log(self.__class__.__name__, 'research_start', {'topic': topic})

        # План исследования
        plan = self._create_research_plan(topic)
        self.logger.log(self.__class__.__name__, 'research_plan', plan)

        results = {}
        for step in plan['steps']:
            try:
                if step['type'] == 'code':
                    result = self.code_agent.generate_and_execute(step['task'])
                    results[step['id']] = result
                elif step['type'] == 'analysis':
                    analysis = self._perform_analysis(step['task'], results)
                    results[step['id']] = analysis
            except Exception as e:
                self.logger.log(self.__class__.__name__, 'error',
                              {'step': step, 'error': str(e)})

        conclusion = self._generate_conclusion(topic, results)
        self.logger.log(self.__class__.__name__, 'research_complete', conclusion)
        return conclusion

    def _create_research_plan(self, topic: str) -> dict:
        prompt = f"""Создай план исследования для темы: {topic}
        Формат:
        - Типы шагов: code, analysis
        - Каждый шаг должен иметь id, описание и тип"""
        response = self.llm.generate_response(prompt)
        return self._parse_plan(response)

    def _perform_analysis(self, task: str, data: dict) -> dict:
        prompt = f"""Проанализируй данные: {json.dumps(data)}
        Задача анализа: {task}
        Ответ в формате JSON"""
        return json.loads(self.llm.generate_response(prompt))

    def _generate_conclusion(self, topic: str, results: dict) -> dict:
        prompt = f"""Сгенерируй заключение по теме {topic} на основе результатов:
        {json.dumps(results)}
        Ответ на русском языке в формате Markdown"""
        return {'conclusion': self.llm.generate_response(prompt)}

# Пример использования
if __name__ == "__main__":
    logger = EnhancedLogger()
    llm = together.Together()

    research_agent = ResearchAgent(llm, logger)

    result = research_agent.conduct_research(
        "Анализ рынка AI-чипов: сравнение производительности, "
        "расчет стоимости вычислений и прогноз трендов"
    )

    print("Результат исследования:")
    print(json.dumps(result, indent=2, ensure_ascii=False))

    # Генерация отчета
    with open(f"report_{logger.session_id}.md", 'w') as f:
        f.write(result['conclusion'])
```

```python
import together
import datetime
import json
import numpy as np
import uuid
from functools import lru_cache
from sentence_transformers import SentenceTransformer
from typing import List, Dict, Tuple, Optional

# Инициализация моделей
encoder_model = SentenceTransformer('paraphrase-multilingual-mpnet-base-v2')

# ==================== Базовые классы ====================
class DeepSeekAgent:
    """Класс для взаимодействия с LLM через Together API"""
    def __init__(self, model_name: str = "deepseek-ai/DeepSeek-R1-Distill-Llama-70B-free"):
        self.client = together.Together()
        self.model = model_name

    @lru_cache(maxsize=512)
    def generate_response(self, user_input: str, max_tokens: int = 512, temperature: float = 0.7) -> str:
        try:
            response = self.client.chat.completions.create(
                model=self.model,
                messages=[{"role": "user", "content": user_input}],
                max_tokens=max_tokens,
                temperature=temperature
            )
            return response.choices[0].message.content
        except Exception as e:
            return f"Ошибка LLM: {str(e)}"

class Tool:
    """Класс для определения инструментов"""
    def __init__(self, name: str, func: callable, description: str):
        self.name = name
        self.func = func
        self.description = description

class ExecutionLogger:
    """Расширенная система логирования выполнения задач"""
    def __init__(self, session_id: str = None):
        self.session_id = session_id or str(uuid.uuid4())
        self.filename = f"execution_log_{self.session_id}.jsonl"
        self.log_buffer = []

    def log_event(self, event_type: str, agent: str, details: dict) -> None:
        entry = {
            "timestamp": datetime.datetime.now().isoformat(),
            "session_id": self.session_id,
            "event_type": event_type,
            "agent": agent,
            **details
        }
        self.log_buffer.append(entry)
        self._flush_to_file()

    def _flush_to_file(self) -> None:
        with open(self.filename, "a", encoding="utf-8") as f:
            for entry in self.log_buffer:
                f.write(json.dumps(entry, ensure_ascii=False) + "\n")
            self.log_buffer.clear()

    def generate_report(self) -> str:
        """Генерация текстового отчета из логов"""
        report = []
        with open(self.filename, "r") as f:
            logs = [json.loads(line) for line in f]

        timeline = []
        agents = {}
        errors = []

        for log in logs:
            timeline.append(f"{log['timestamp']} - {log['event_type']} ({log['agent']})")

            if log['agent'] not in agents:
                agents[log['agent']] = {
                    "tasks": 0,
                    "tools_used": 0,
                    "errors": 0
                }

            agents[log['agent']]["tasks"] += 1

            if "error" in log:
                errors.append(log)
                agents[log['agent']]["errors"] += 1

            if log['event_type'] == "TOOL_EXECUTION_SUCCESS":
                agents[log['agent']]["tools_used"] += 1

        report.append("="*40 + " Execution Report " + "="*40)
        report.append(f"Session ID: {self.session_id}")
        report.append(f"Total events: {len(logs)}")
        report.append(f"Errors detected: {len(errors)}")
        report.append("\nTimeline Summary:")
        report.extend(timeline[:5] + ["..."] + timeline[-5:])

        report.append("\nAgent Performance:")
        for agent, stats in agents.items():
            report.append(
                f"{agent}: "
                f"Tasks: {stats['tasks']}, "
                f"Tools Used: {stats['tools_used']}, "
                f"Errors: {stats['errors']}"
            )

        if errors:
            report.append("\nError Details:")
            for error in errors[:3]:
                report.append(f"{error['timestamp']} - {error.get('message', '')}")

        return "\n".join(report)

# ==================== Основные компоненты системы ====================
class Memory:
    """Векторная база знаний с семантическим поиском"""
    def __init__(self):
        self.memories = []
        self.encoder = encoder_model

    def add(self, text: str) -> None:
        embedding = self.encoder.encode(text)
        self.memories.append({'text': text, 'embedding': embedding})

    def search(self, query: str, top_k: int = 3) -> List[Tuple[str, float]]:
        query_emb = self.encoder.encode(query)
        similarities = []
        for mem in self.memories:
            sim = np.dot(query_emb, mem['embedding'])
            similarities.append((mem['text'], sim))
        return sorted(similarities, key=lambda x: x[1], reverse=True)[:top_k]

class SecurityLayer:
    """Система безопасности и контроля контента"""
    def __init__(self):
        self.forbidden_keywords = [
            "пароль", "кредитная карта", "конфиденциально", "секретно"
        ]

    def validate(self, text: str) -> Tuple[bool, str]:
        for keyword in self.forbidden_keywords:
            if keyword in text.lower():
                return False, f"Обнаружено запрещенное слово: {keyword}"
        return True, "OK"

class Agent:
    """Базовый класс интеллектуального агента"""
    def __init__(self,
               name: str,
               llm: Optional[DeepSeekAgent] = None,
               tools: Optional[List[Tool]] = None,
               logger: Optional[ExecutionLogger] = None):
        self.name = name
        self.llm = llm or DeepSeekAgent()
        self.memory = Memory()
        self.tools = {t.name: t for t in (tools or [])}
        self.security = SecurityLayer()
        self.logger = logger or ExecutionLogger()
        self.skills = {}

    def add_tool(self, tool: Tool) -> None:
        self.tools[tool.name] = tool

    def add_skill(self, skill: str, confidence: float) -> None:
        self.skills[skill] = confidence

    def evaluate_competence(self, task: str) -> float:
        prompt = f"""Оцени свою способность решить задачу от 0 до 1. Ответь только числом.
        Задача: {task}
        Навыки: {', '.join(self.skills.keys())}"""
        try:
            return float(self.llm.generate_response(prompt))
        except:
            return 0.0

    def solve_task(self, task: str) -> str:
        self.logger.log_event("TASK_RECEIVED", self.name, {"task": task})

        try:
            # Проверка безопасности
            is_valid, msg = self.security.validate(task)
            if not is_valid:
                self.logger.log_event("SECURITY_BLOCK", self.name, {"message": msg})
                return msg

            # Поиск в памяти
            similar = self.memory.search(task)
            if similar and similar[0][1] > 0.85:
                self.logger.log_event("MEMORY_USED", self.name, {"content": similar[0][0]})
                return similar[0][0]

            # Использование инструментов
            if self.tools:
                tool_list = [f"{t.name} ({t.description})" for t in self.tools.values()]
                tool_prompt = f"""Доступные инструменты: {tool_list}
                Задача: {task}
                Выбери инструмент и аргументы в формате: ИНСТРУМЕНТ|АРГУМЕНТЫ"""

                self.logger.log_event("TOOL_SELECTION_START", self.name, {})
                tool_response = self.llm.generate_response(tool_prompt)
                self.logger.log_event("TOOL_SELECTION_RESULT", self.name,
                                    {"response": tool_response})

                if '|' in tool_response:
                    tool_name, args = tool_response.split('|', 1)
                    tool_name = tool_name.strip()
                    args = args.strip()

                    if tool_name in self.tools:
                        self.logger.log_event("TOOL_EXECUTION_START", self.name,
                                           {"tool": tool_name, "args": args})
                        result = self.tools[tool_name].func(args)
                        self.memory.add(f"Задача: {task}\nРешение: {result}")
                        self.logger.log_event("TOOL_EXECUTION_SUCCESS", self.name,
                                           {"tool": tool_name, "result": result})
                        return result
                    else:
                        error_msg = f"Инструмент {tool_name} не найден"
                        self.logger.log_event("TOOL_ERROR", self.name,
                                           {"error": error_msg})
                        return error_msg

            # Решение через LLM
            self.logger.log_event("LLM_PROCESSING_START", self.name, {"prompt": task})
            result = self.llm.generate_response(task)
            self.memory.add(f"Задача: {task}\nРешение: {result}")
            self.logger.log_event("LLM_PROCESSING_END", self.name, {"result": result})

            return result

        except Exception as e:
            error_msg = f"Критическая ошибка: {str(e)}"
            self.logger.log_event("TASK_FAILED", self.name, {"error": error_msg})
            return error_msg

class BossAgent(Agent):
    """Координатор задач с расширенными возможностями"""
    def __init__(self,
               name: str,
               team: List[Agent],
               **kwargs):
        super().__init__(name, **kwargs)
        self.team = team
        self.decomposer = TaskDecomposer(self.llm)
        self.verifier = Agent("Verifier", llm=self.llm, logger=self.logger)

    def solve_task(self, task: str, retries: int = 2) -> str:
        self.logger.log_event("TASK_START", self.name, {"task": task, "retries": retries})

        try:
            # Разбиение задачи
            self.logger.log_event("TASK_DECOMPOSITION_START", self.name, {})
            subtasks = self.decomposer.decompose(task)
            self.logger.log_event("TASK_DECOMPOSITION_END", self.name,
                                {"subtasks": subtasks})

            results = {}
            for key, subtask in subtasks.items():
                # Выбор агента
                self.logger.log_event("AGENT_SELECTION_START", self.name,
                                    {"subtask_id": key})
                agent = self.select_agent(subtask)
                self.logger.log_event("AGENT_SELECTED", self.name,
                                    {"subtask_id": key, "agent": agent.name})

                # Выполнение подзадачи
                self.logger.log_event("SUBTASK_START", agent.name,
                                    {"subtask_id": key, "description": subtask})
                result = agent.solve_task(subtask)
                self.logger.log_event("SUBTASK_COMPLETE", agent.name,
                                    {"subtask_id": key, "result": result})
                results[key] = result

            # Агрегация результатов
            combined = "\n".join([f"{k}: {v}" for k, v in results.items()])
            self.logger.log_event("RESULT_AGGREGATION", self.name,
                                {"combined_result": combined})

            # Верификация
            self.logger.log_event("VERIFICATION_START", self.name, {})
            verification = self.verifier.solve_task(f"Проверь: {combined}")
            self.logger.log_event("VERIFICATION_RESULT", self.name,
                                {"verification_result": verification})

            # Обработка ошибок
            if "ошибка" in verification.lower() and retries > 0:
                self.logger.log_event("RETRY_INITIATED", self.name,
                                    {"retries_left": retries-1})
                return self._retry_mechanism(task, retries)

            final_result = f"Финальный отчет:\n{combined}\nПроверка: {verification}"
            self.logger.log_event("TASK_COMPLETE", self.name,
                                {"final_result": final_result})
            return final_result

        except Exception as e:
            error_msg = f"Системная ошибка: {str(e)}"
            self.logger.log_event("TASK_FAILED", self.name, {"error": error_msg})
            return error_msg

    def select_agent(self, subtask: str) -> Agent:
        scores = {agent.name: agent.evaluate_competence(subtask) for agent in self.team}
        return max(self.team, key=lambda a: scores[a.name])

    def _retry_mechanism(self, task: str, retries: int) -> str:
        self.team = self.team[1:] + self.team[:1]  # Ротация агентов
        return self.solve_task(task, retries-1)

class TaskDecomposer:
    """Мультистратегическое разбиение задач"""
    STRATEGIES = {
        'sequential': "Разбей на последовательные шаги :",
        'parallel': "Разбей на независимые подзадачи:",
        'expert': "Определи необходимые области экспертизы:"
    }

    def __init__(self, llm: DeepSeekAgent):
        self.llm = llm

    def decompose(self, task: str, strategy: str = 'auto') -> Dict[str, str]:
        if strategy == 'auto':
            prompt = f"Выбери лучший метод разбиения (sequential/parallel/expert): {task}"
            strategy = self.llm.generate_response(prompt).lower()[:3]

        prompt = f"{self.STRATEGIES.get(strategy, 'Разбей на подзадачи:')}\n{task}"
        response = self.llm.generate_response(prompt)
        return self._parse_subtasks(response)

    def _parse_subtasks(self, text: str) -> Dict[str, str]:
        tasks = {}
        for line in text.split('\n'):
            if '.' in line:
                num, desc = line.split('.', 1)
                tasks[num.strip()] = desc.strip()
        return tasks
```

```python

# ==================== Пример использования ====================
if __name__ == "__main__":
    tools = []    # Инициализация системы
    session_logger = ExecutionLogger()
    llm = DeepSeekAgent()

    # Создание команды агентов
    analyst = Agent(
        name="Програмист",
        llm=llm,
        tools=tools,
        logger=session_logger
    )
    analyst.add_skill("Хорошо разбирается в нейросетях", 0.9)

    engineer = Agent(
        name="Аналитик",
        llm=llm,
        tools=tools,
        logger=session_logger
    )
    engineer.add_skill("Умеет анализировать и экспериментировать ", 0.95)

    researcher = Agent(
        name="кодер",
        llm=llm,
        tools=tools,
        logger=session_logger
    )
    researcher.add_skill("хорошо пишет код", 0.85)

    # Создание главного агента
    boss = BossAgent(
        name="Главный координатор",
        team=[analyst, engineer, researcher],
        llm=llm,
        logger=session_logger
    )

    # Пример комплексной задачи
    task = """
    Реализуй код нейросети для игры в дилему заключённого. Идея: нужно решить задачу на подобии задачи "дилемма заключённого" . Суть проста , два юнита начинают взаимодействовать друг с другом N раз , за одно взаимодействие они могут 1) подставить другого или 2) довериться другому . Результат одного действия оценивается следующим образом : Если оба доверились , то каждый получает по 3 очка , если оба подставили , то каждый получает 1 очков , если один доверился а другой его подставил , то подставивший получает 5 очков , а доверившийся 0. Суть набрать как можно больше очков
    """

    # Выполнение задачи
    print("Запуск выполнения задачи...\n")
    result = boss.solve_task(task)

    # Вывод результатов
    print("="*60)
    print("Финальный результат:")
    print(result)

    # Генерация отчета
    report = session_logger.generate_report()
    print("\n" + "="*60)
    print("Отчет о выполнении:")
    print(report)
    print(f"\nПолный лог выполнения сохранен в файл: {session_logger.filename}")
```

```python
# @title json распаковка



import json
import textwrap

def format_entry(entry, words_per_line=10):
    """Форматирует одну запись из лога"""
    formatted = []

    # Основные поля
    formatted.append(f"Время: {entry.get('timestamp', 'N/A')}")
    formatted.append(f"Агент: {entry.get('agent', 'N/A')}")
    formatted.append(f"Тип события: {entry.get('event_type', 'N/A')}")
    formatted.append(f"Подзадача: {entry.get('subtask_id', 'N/A')}")

    # Обработка результата с переносами
    if 'result' in entry:
        words = entry['result'].split()
        chunks = [words[i:i+words_per_line]
                  for i in range(0, len(words), words_per_line)]

        formatted.append("\nРезультат:")
        for chunk in chunks:
            formatted.append("  " + " ".join(chunk))

    return "\n".join(formatted)

def convert_jsonl_to_text(input_file, output_file, words_per_line=10):
    """Конвертирует JSONL в текстовый формат"""
    with open(input_file, 'r', encoding='utf-8') as infile, \
         open(output_file, 'w', encoding='utf-8') as outfile:

        for line in infile:
            try:
                entry = json.loads(line)
                formatted = format_entry(entry, words_per_line)
                outfile.write(formatted + "\n")
                outfile.write("-" * 80 + "\n\n")
            except json.JSONDecodeError:
                print(f"Ошибка в строке: {line}")

# Использование
convert_jsonl_to_text(
    input_file=session_logger.filename,
    output_file='formatted_log.txt',
    words_per_line=10
)
```

```python
import together
import datetime
import json
import numpy as np
import uuid
from functools import lru_cache
from sentence_transformers import SentenceTransformer
from typing import List, Dict, Tuple, Optional

# Инициализация модели эмбеддингов
encoder_model = SentenceTransformer('paraphrase-multilingual-mpnet-base-v2')

# ==================== Классы для взаимодействия с LLM ====================

class DeepSeekAgent:
    """Интеграция с LLM через Together API."""
    def __init__(self, model_name: str = "deepseek-ai/DeepSeek-R1-Distill-Llama-70B-free"):
        self.client = together.Together()
        self.model = model_name

    @lru_cache(maxsize=512)
    def generate_response(self, user_input: str, max_tokens: int = 512, temperature: float = 0.7) -> str:
        try:
            response = self.client.chat.completions.create(
                model=self.model,
                messages=[{"role": "user", "content": user_input}],
                max_tokens=max_tokens,
                temperature=temperature
            )
            return response.choices[0].message.content
        except Exception as e:
            return f"Ошибка LLM: {str(e)}"

# ==================== Дополнительные классы ====================

class Tool:
    """Класс описания инструмента."""
    def __init__(self, name: str, func: callable, description: str):
        self.name = name
        self.func = func
        self.description = description

class ExecutionLogger:
    """Система логирования и генерации отчётов выполнения."""
    def __init__(self, session_id: str = None):
        self.session_id = session_id or str(uuid.uuid4())
        self.filename = f"execution_log_{self.session_id}.jsonl"
        self.log_buffer = []

    def log_event(self, event_type: str, agent: str, details: dict) -> None:
        entry = {
            "timestamp": datetime.datetime.now().isoformat(),
            "session_id": self.session_id,
            "event_type": event_type,
            "agent": agent,
            **details
        }
        self.log_buffer.append(entry)
        self._flush_to_file()

    def _flush_to_file(self) -> None:
        with open(self.filename, "a", encoding="utf-8") as f:
            for entry in self.log_buffer:
                f.write(json.dumps(entry, ensure_ascii=False) + "\n")
            self.log_buffer.clear()

    def generate_report(self) -> str:
        report = []
        with open(self.filename, "r", encoding="utf-8") as f:
            logs = [json.loads(line) for line in f]

        timeline = []
        agents = {}
        errors = []

        for log in logs:
            timeline.append(f"{log['timestamp']} - {log['event_type']} ({log['agent']})")
            if log['agent'] not in agents:
                agents[log['agent']] = {"tasks": 0, "tools_used": 0, "errors": 0}
            agents[log['agent']]["tasks"] += 1
            if "error" in log:
                errors.append(log)
                agents[log['agent']]["errors"] += 1
            if log['event_type'] == "TOOL_EXECUTION_SUCCESS":
                agents[log['agent']]["tools_used"] += 1

        report.append("=" * 40 + " Execution Report " + "=" * 40)
        report.append(f"Session ID: {self.session_id}")
        report.append(f"Total events: {len(logs)}")
        report.append(f"Errors detected: {len(errors)}")
        report.append("\nTimeline Summary:")
        report.extend(timeline[:5] + ["..."] + timeline[-5:])
        report.append("\nAgent Performance:")
        for agent, stats in agents.items():
            report.append(
                f"{agent}: Tasks: {stats['tasks']}, Tools Used: {stats['tools_used']}, Errors: {stats['errors']}"
            )
        if errors:
            report.append("\nError Details:")
            for error in errors[:3]:
                report.append(f"{error['timestamp']} - {error.get('message', '')}")
        return "\n".join(report)

# ==================== Компоненты системы ====================

class Memory:
    """Векторная база знаний для семантического поиска."""
    def __init__(self):
        self.memories = []
        self.encoder = encoder_model

    def add(self, text: str) -> None:
        embedding = self.encoder.encode(text)
        self.memories.append({'text': text, 'embedding': embedding})

    def search(self, query: str, top_k: int = 3) -> List[Tuple[str, float]]:
        query_emb = self.encoder.encode(query)
        similarities = []
        for mem in self.memories:
            sim = np.dot(query_emb, mem['embedding'])
            similarities.append((mem['text'], sim))
        return sorted(similarities, key=lambda x: x[1], reverse=True)[:top_k]

class SecurityLayer:
    """Проверка входящих данных на наличие запрещённых слов."""
    def __init__(self):
        self.forbidden_keywords = ["пароль", "кредитная карта", "конфиденциально", "секретно"]

    def validate(self, text: str) -> Tuple[bool, str]:
        for keyword in self.forbidden_keywords:
            if keyword in text.lower():
                return False, f"Обнаружено запрещенное слово: {keyword}"
        return True, "OK"

# ==================== Основной класс агента ====================

class Agent:
    """Базовый интеллектуальный агент с улучшенной интеграцией LLM."""
    def __init__(self,
                 name: str,
                 llm: Optional[DeepSeekAgent] = None,
                 tools: Optional[List[Tool]] = None,
                 logger: Optional[ExecutionLogger] = None):
        self.name = name
        self.llm = llm or DeepSeekAgent()
        self.memory = Memory()
        self.tools = {t.name: t for t in (tools or [])}
        self.security = SecurityLayer()
        self.logger = logger or ExecutionLogger()
        self.skills = {}

    def add_tool(self, tool: Tool) -> None:
        self.tools[tool.name] = tool

    def add_skill(self, skill: str, confidence: float) -> None:
        self.skills[skill] = confidence

    def evaluate_competence(self, task: str, expected: str) -> float:
        prompt = f"""Оцени свою способность решить задачу по следующим данным:
Задача: {task}
Ожидаемый результат: {expected}
Ответь числом от 0 до 1, без пояснений."""
        try:
            response = self.llm.generate_response(prompt)
            return float(response.strip())
        except Exception:
            return 0.0

    def solve_task(self, payload: Dict[str, str]) -> str:
        """
        Ожидаемый формат payload:
          {
              "task": "описание задачи",
              "expected": "ожидаемый результат/критерии успеха"
          }
        """
        task = payload.get("task", "")
        expected_result = payload.get("expected", "")
        self.logger.log_event("TASK_RECEIVED", self.name, {"task": task, "expected": expected_result})

        # Проверка безопасности
        is_valid, msg = self.security.validate(task)
        if not is_valid:
            self.logger.log_event("SECURITY_BLOCK", self.name, {"message": msg})
            return msg

        try:
            # Поиск похожего решения в памяти
            similar = self.memory.search(task)
            if similar and similar[0][1] > 0.85:
                self.logger.log_event("MEMORY_USED", self.name, {"content": similar[0][0]})
                return similar[0][0]

            # Если доступны инструменты, спросим LLM в структурированном формате
            if self.tools:
                tool_dict = {t.name: t.description for t in self.tools.values()}
                tool_prompt = f"""У тебя есть следующие доступные инструменты:
{json.dumps(tool_dict, ensure_ascii=False, indent=2)}
Задача: {task}
Ожидаемый результат: {expected_result}
Пожалуйста, выбери инструмент и аргументы в формате JSON: {{"tool": "название_инструмента", "args": "аргументы"}}"""
                self.logger.log_event("TOOL_SELECTION_START", self.name, {})
                tool_response = self.llm.generate_response(tool_prompt)
                self.logger.log_event("TOOL_SELECTION_RESULT", self.name, {"response": tool_response})
                try:
                    tool_data = json.loads(tool_response)
                    tool_name = tool_data.get("tool", "").strip()
                    args = tool_data.get("args", "").strip()
                except Exception:
                    tool_name = ""
                    args = ""
                if tool_name in self.tools:
                    self.logger.log_event("TOOL_EXECUTION_START", self.name, {"tool": tool_name, "args": args})
                    result = self.tools[tool_name].func(args)
                    self.memory.add(f"Задача: {task}\nРешение: {result}")
                    self.logger.log_event("TOOL_EXECUTION_SUCCESS", self.name, {"tool": tool_name, "result": result})
                    return result
                elif tool_name:
                    error_msg = f"Инструмент {tool_name} не найден"
                    self.logger.log_event("TOOL_ERROR", self.name, {"error": error_msg})
                    return error_msg

            # Решение через LLM
            self.logger.log_event("LLM_PROCESSING_START", self.name, {"prompt": task})
            initial_result = self.llm.generate_response(task)
            # Запрос на уточнение решения для лучшего соответствия ожидаемому результату
            refinement_prompt = f"""У нас есть первоначальное решение:
{initial_result}

Ожидаемый результат:
{expected_result}

Как можно улучшить начальное решение так, чтобы оно более точно соответствовало ожидаемому результату? Предоставь уточнённый ответ."""
            refined_result = self.llm.generate_response(refinement_prompt)
            # Добавление результата в память
            final_result = refined_result if refined_result.strip() else initial_result
            self.memory.add(f"Задача: {task}\nРешение: {final_result}")
            self.logger.log_event("LLM_PROCESSING_END", self.name, {"result": final_result})
            return final_result

        except Exception as e:
            error_msg = f"Критическая ошибка: {str(e)}"
            self.logger.log_event("TASK_FAILED", self.name, {"error": error_msg})
            return error_msg

# ==================== Компонент для декомпозиции задач ====================

class TaskDecomposer:
    """Мультистратегическое разбиение задач с учётом ожидаемого результата."""
    STRATEGIES = {
        'sequential': "Разбей задачу на последовательные шаги:",
        'parallel': "Разбей задачу на независимые подзадачи:",
        'expert': "Определи необходимые области экспертизы для решения задачи:"
    }

    def __init__(self, llm: DeepSeekAgent):
        self.llm = llm

    def decompose(self, task: str, expected: str, strategy: str = 'auto') -> Dict[str, str]:
        if strategy == 'auto':
            prompt = f"Выбери лучший метод разбиения задачи (sequential/parallel/expert) с учётом ожидаемого результата.\nЗадача: {task}\nОжидаемый результат: {expected}"
            strategy_response = self.llm.generate_response(prompt)
            # Берем первые три символа (например, "seq", "par", "exp"), но можно расширить логику
            strategy = strategy_response.lower()[:3]

        base_instruction = self.STRATEGIES.get(strategy, "Разбей задачу на подзадачи:")
        prompt = f"{base_instruction}\nЗадача: {task}\nОжидаемый результат: {expected}"
        response = self.llm.generate_response(prompt)
        return self._parse_subtasks(response)

    def _parse_subtasks(self, text: str) -> Dict[str, str]:
        tasks = {}
        for line in text.split('\n'):
            if '.' in line:
                num, desc = line.split('.', 1)
                tasks[num.strip()] = desc.strip()
        return tasks

# ==================== Расширенный агент-босс ====================

class BossAgent(Agent):
    """Координатор задач с декомпозицией, распределением и итеративной верификацией."""
    def __init__(self,
                 name: str,
                 team: List[Agent],
                 **kwargs):
        super().__init__(name, **kwargs)
        self.team = team
        self.decomposer = TaskDecomposer(self.llm)
        self.verifier = Agent("Verifier", llm=self.llm, logger=self.logger)

    def solve_task(self, payload: Dict[str, str], retries: int = 2) -> str:
        task = payload.get("task", "")
        expected_result = payload.get("expected", "")
        self.logger.log_event("TASK_START", self.name, {"task": task, "expected": expected_result, "retries": retries})
        try:
            # Декомпозиция задачи с учётом ожидаемого результата
            self.logger.log_event("TASK_DECOMPOSITION_START", self.name, {})
            subtasks = self.decomposer.decompose(task, expected_result)
            self.logger.log_event("TASK_DECOMPOSITION_END", self.name, {"subtasks": subtasks})
            results = {}
            # Решение каждой подзадачи выбранным агентом
            for key, subtask in subtasks.items():
                self.logger.log_event("AGENT_SELECTION_START", self.name, {"subtask_id": key})
                agent = self.select_agent(subtask, expected_result)
                self.logger.log_event("AGENT_SELECTED", self.name, {"subtask_id": key, "agent": agent.name})
                self.logger.log_event("SUBTASK_START", agent.name, {"subtask_id": key, "description": subtask})
                # Каждый агент получает подзадачу с ожидаемым результатом
                subpayload = {"task": subtask, "expected": expected_result}
                result = agent.solve_task(subpayload)
                self.logger.log_event("SUBTASK_COMPLETE", agent.name, {"subtask_id": key, "result": result})
                results[key] = result

            # Агрегация результатов подзадач
            combined = "\n".join([f"{k}: {v}" for k, v in results.items()])
            self.logger.log_event("RESULT_AGGREGATION", self.name, {"combined_result": combined})
            # Верификация агрегированного результата
            self.logger.log_event("VERIFICATION_START", self.name, {})
            verification = self.verifier.solve_task({"task": f"Проверь итоговое решение:\n{combined}", "expected": expected_result})
            self.logger.log_event("VERIFICATION_RESULT", self.name, {"verification_result": verification})
            # Если результат не соответствует ожиданиям и есть попытки, выполнить ретрай
            if "ошибка" in verification.lower() and retries > 0:
                self.logger.log_event("RETRY_INITIATED", self.name, {"retries_left": retries-1})
                return self._retry_mechanism(payload, retries)
            final_result = f"Финальный отчет:\n{combined}\nПроверка: {verification}"
            self.logger.log_event("TASK_COMPLETE", self.name, {"final_result": final_result})
            return final_result

        except Exception as e:
            error_msg = f"Системная ошибка: {str(e)}"
            self.logger.log_event("TASK_FAILED", self.name, {"error": error_msg})
            return error_msg

    def select_agent(self, subtask: str, expected: str) -> Agent:
        # Оценка компетентности с учётом ожидаемого результата
        scores = {agent.name: agent.evaluate_competence(subtask, expected) for agent in self.team}
        return max(self.team, key=lambda a: scores[a.name])

    def _retry_mechanism(self, payload: Dict[str, str], retries: int) -> str:
        # Простая ротация агентов для смены подхода
        self.team = self.team[1:] + self.team[:1]
        return self.solve_task(payload, retries-1)

# ==================== Пример использования ====================


  # Пример функции для инструмента (например, калькуляция)
tools = [
Tool("Калькулятор", calculate, "Математические вычисления"),
Tool("Веб-поиск", web_search, "Поиск информации в интернете"),
Tool("Запуск кода", colab_exec, "Запускает и редактирует код в Google Colab")
]
    # Инициализируем базовых агентов с инструментами
agent1 = Agent(name="Agent1", tools=tools)
agent2 = Agent(name="Agent2", tools=tools)
team = [agent1, agent2]

    # Инициализируем BossAgent
boss = BossAgent(name="Boss", team=team)

    # Пример входного payload
payload = {
        "task": "Прибумать юмаристическую песню про древних русов и ящеров , Более того, древние русы и динозавры познакомились ещё 68 миллионов лет тому назад! При этом динозавры настолько преисполнились в познании веры русов, что при своём царе Тираннозавре IV приняли славянское язычество. Ну а означенный Тираннозавр IV переименовался в Тираннослава I. После чего динозавры стали помогать древним русам отгонять от их земель… монгольские орды. Которые, видимо, тоже существовали уже 68 миллионов лет назад.",
        "expected": "Ожидается получить текст песни в размере 3 куплетов"
}

    # Выполнение задачи через BossAgent
final_output = boss.solve_task(payload)
print("Итоговый результат работы BossAgent:")
print(final_output)
```

```python
import requests
import xml.etree.ElementTree as ET

# Укажите ваш Wolfram Alpha App ID (полученный после регистрации)
WOLFRAM_APP_ID = "9QJQ4G-4735V7VH7E"

def calculate(expression: str) -> str:
    """
    Расширенная функция вычислений:
      - Если задан Wolfram Alpha App ID, выполняется запрос к Wolfram Alpha.
      - Если API недоступен или не задан, производится локальный расчет с проверкой безопасности.
    """
    # Если есть API-ключ для Wolfram Alpha и запрос не слишком короткий (например, имеет смысл для более сложных задач)
    if WOLFRAM_APP_ID and len(expression.strip()) > 0:
        url = "http://api.wolframalpha.com/v2/query"
        params = {
            "appid": WOLFRAM_APP_ID,
            "input": expression,
            "output": "XML"
        }
        try:
            response = requests.get(url, params=params, timeout=10)
            if response.status_code == 200:
                # Парсим XML-ответ
                root = ET.fromstring(response.text)
                # Ищем pod с id 'Result', в котором обычно содержится основной результат
                for pod in root.findall(".//pod[@id='Result']"):
                    # Берем первый plaintext
                    plaintext = pod.find("subpod/plaintext")
                    if plaintext is not None and plaintext.text:
                        return plaintext.text.strip()
                # Если не нашли result, можно попытаться взять другой pod, например 'Input interpretation'
                return "Wolfram Alpha не смог дать однозначный ответ."
            else:
                return f"Ошибка Wolfram Alpha: HTTP {response.status_code}"
        except Exception as e:
            # Если возникла ошибка, можно вернуться к локальному вычислению
            return f"Ошибка Wolfram Alpha: {str(e)}. Пытаемся локально."

    # Локальное вычисление с проверкой допустимых символов (на случай, если API не используется или не отвечает)
    allowed_chars = "0123456789+-*/(). "
    if not all(c in allowed_chars for c in expression):
        return "Выражение содержит недопустимые символы."
    try:
        result = eval(expression, {"__builtins__": {}})
        return str(result)
    except Exception as e:
        return f"Ошибка вычисления: {str(e)}"
```

```python
from bs4 import BeautifulSoup
import requests

def get_page_title(url: str) -> str:
    try:
        response = requests.get(url, timeout=10)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, "html.parser")
            title = soup.find("title")
            return title.text.strip() if title else "Нет заголовка"
        else:
            return f"Ошибка загрузки страницы: HTTP {response.status_code}"
    except Exception as e:
        return f"Ошибка загрузки страницы: {str(e)}"

def web_search(query: str) -> str:
    """
    Поиск информации в интернете с использованием googlesearch-python.
    В качестве расширения можно извлечь заголовки страниц через BeautifulSoup.
    """
    try:
        from googlesearch import search
        results = list(search(query, num_results=3))
        if not results:
            return "По вашему запросу ничего не найдено."

        output_lines = []
        for url in results:
            title = get_page_title(url)
            output_lines.append(f"{title}\n{url}")
        return "Найденные ссылки:\n" + "\n\n".join(output_lines)
    except Exception as e:
        return f"Ошибка веб-поиска: {str(e)}"
```

```python
import datetime
import json
import numpy as np
import uuid
from functools import lru_cache
from sentence_transformers import SentenceTransformer
from typing import List, Dict, Tuple, Optional


def colab_exec(code: str) -> str:
    """
    Функция для выполнения кода в Google Colab:
      - Выполняет переданный код, захватывает вывод.
      - Затем устанавливает тот же код в следующую ячейку Colab для последующего редактирования.
    """
    import sys
    from io import StringIO

    # Захват вывода
    old_stdout = sys.stdout
    sys.stdout = mystdout = StringIO()
    try:
        # Безопасное выполнение кода с глобальным пространством имен
        exec(code, globals())
        result = mystdout.getvalue()
    except Exception as e:
        result = f"Ошибка выполнения кода: {str(e)}"
    finally:
        sys.stdout = old_stdout

    # Если выполняемся в Jupyter/Colab, устанавливаем переданный код как содержимое следующей ячейки
    try:
        from IPython import get_ipython
        ip = get_ipython()
        if ip:
            # Устанавливаем следующий ввод без перезаписи текущей ячейки (replace=False)
            ip.set_next_input(code, replace=False)
    except Exception:
        pass

    return result if result else "Код выполнен. Следующий блок создан для редактирования."
```
